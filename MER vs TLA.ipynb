# Comparative Time Series Analysis of Air Quality at MER and TLA Stations

## Overall Trends in AQI

**Similarities:**

* **General Trends:** Both stations exhibit similar overall patterns in their monthly average AQI values. AQI levels drop noticeably around mid-2022, recover, and then decline again toward late 2023. This suggests that both locations are influenced by regional factors influencing air quality.

* **Volatility:** Both stations exhibit significant daily fluctuations in AQI, indicating that day-to-day changes in weather, pollution sources, and human activity impact air quality in both areas.

**Differences:**

* **Absolute AQI Values:** The TLA station consistently shows lower AQI values compared to the MER station, suggesting that TLA generally experiences better air quality than MER. The difference is notable, approximately 0.5 on the AQI scale.

* **Magnitude of Fluctuations:** While both stations have daily variability, the range of fluctuations appears slightly larger at the MER station. This could imply that MER is more susceptible to sudden spikes in pollution levels.

* **Specific Peaks and Dips:** While the overall trends are similar, the exact timing and magnitude of peaks and dips in the monthly averages don't perfectly align. This indicates that local factors also play a role in air quality at each station. For instance, the sharp increase in AQI at MER towards the end of the time series is not mirrored as strongly at TLA.


## Smoothed AQI (30-Day Rolling Average)

**Similarities:**

* **General Trends (Revealed by Rolling Average):** The 30-day rolling averages for both TLA and MER reveal broadly similar long-term trends. Both show a dip in AQI around mid-2022, indicating a period of improved air quality sustained over a longer timescale. Both also show an increase in AQI towards the end of 2023, suggesting a decline in air quality that persists beyond short-term fluctuations. These similar trends in the smoothed data strongly suggest that both stations are influenced by larger regional influences on air quality.

**Differences:**

* **Absolute AQI Values:** The 30-day rolling average clearly shows that TLA consistently maintains lower AQI values than MER throughout the entire period. This difference reflects a sustained difference in air quality between the two locations, as evidenced by the smoothed data.

* **Magnitude of Fluctuations:** While the 30-day rolling average smooths out daily variations, the magnitude of the remaining fluctuations in the smoothed data differs between the two stations. MER exhibits larger swings in its 30-day rolling average than TLA. This suggests that MER might be more susceptible to variations in air quality that persist for longer periods (weeks to months), even after accounting for daily noise.

* **Timing and Amplitude of Peaks/Dips:** While the rolling averages reveal shared general trends, the precise timing and amplitude of some peaks and dips in the smoothed data differ between TLA and MER. For example, while both stations show a dip in mid-2022, the depth and duration of that dip might vary. Similarly, the increase in AQI towards the end of 2023 is considerably larger at MER than at TLA in the smoothed data. These differences in the smoothed data suggest that local factors are influencing air quality at each station on timescales longer than a single day.

**Summary:**

The 30-day rolling averages provide a clearer picture of the underlying air quality trends at TLA and MER by filtering out short-term noise. Similar long-term trends point to regional influences. However, the differences in absolute AQI values, the magnitude of fluctuations, and the timing/amplitude of specific peaks and dips strongly suggest the influence of local factors on timescales longer than a day.

## Air Quality Distribution by Season

* **Overall AQI Levels:** TLA consistently exhibits lower AQI values compared to MER in all seasons, indicating better overall air quality throughout the year. Both MER and TLA stations experience their best air quality during the summer season.

* **Seasonal Variation:** The magnitude of seasonal variation in AQI is greater at MER compared to TLA. This suggests that MER's air quality is more sensitive to seasonal changes in weather patterns and emission sources.

* **Primary Pollutants across Seasons:**
    * **PM2.5 & PM10:** Higher in spring and winter at both stations, likely due to seasonal factors like heating and stagnant atmospheric conditions.
    * **O3:** Peaks in spring at both stations, likely due to photochemical reactions driven by increased sunlight and warmer temperatures.
    * **NO2:** Generally higher in fall and winter at both stations, potentially due to increased emissions from vehicles and combustion during colder months.

**Similarities:**

* **PM2.5 and Ozone:** The contributions of PM2.5 and ozone are roughly similar between the two stations. This indicates that the levels of these pollutants are relatively comparable in both locations.

**Differences:**

## Monthly and Hourly AQI

**Similarities:**

* **Seasonal AQI Trends:** Both stations show higher AQI values in winter (December to February) and lower in summer (June to August), indicating regional influence on air quality.

* **Daily AQI Peak:** Both experience peak AQI in the latter half of the day, between noon and midnight, suggesting human activity's impact on pollution levels.

**Differences:**

* **AQI Magnitude:** MER consistently has higher AQI values than TLA, indicating generally worse air quality. This suggests greater influence from local pollution sources or less favorable conditions for pollutant dispersion at MER.

* **Monthly AQI Pattern:** MER has a more pronounced peak between December and April, with AQI values above 4.5. TLA shows a more gradual increase, peaking around 4.0 in November and December, with further increases between January and May. This difference may reflect variations in local pollution sources or meteorological conditions. For example, MER might have higher winter emissions due to increased industrial activity or reliance on heating sources that produce more particulate matter. Additionally, MER might experience less favorable dispersion conditions, such as temperature inversions or stagnant air masses, which trap pollutants near the ground. On the other hand, TLA might have stricter emission controls or a different mix of pollution sources that contribute less to AQI.

* **Hourly AQI Pattern:** MER's highest average AQI occurs between 12:00 and 15:00. TLA's peak AQI occurs between 17:00 and 19:00 from January to April, and between 13:00 and 15:00 in May. This variation suggests differing local factors influencing air quality. For example, MER might experience peak traffic or industrial activity during midday, leading to higher AQI values. In contrast, TLA's later peak AQI might be influenced by factors such as evening rush hour traffic or delayed dispersion of pollutants due to local wind patterns.

## Conclusion

The comparison of MER and TLA stations highlights both regional influences and local air quality variations. While both stations exhibit similar long-term AQI trends, MER consistently experiences higher pollution levels and greater fluctuations, likely due to stronger local emission sources and less favorable dispersion conditions. Seasonal trends indicate MER’s AQI is more sensitive to weather patterns, while TLA shows a more distinct seasonal cycle. PM10 remains the dominant pollutant at both locations, with temperature and wind speed strongly correlating with AQI variations. Differences in peak AQI hours suggest human activity and atmospheric conditions play a crucial role. Addressing these disparities requires targeted mitigation strategies considering both regional and localized pollution factors.
